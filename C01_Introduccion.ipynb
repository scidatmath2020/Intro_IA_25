## Tipos de IA

Existen tres tipos de inteligencia artificial, es decir, tres categorías a las que comúnmente se refiere la industria. En un extremo del espectro de la IA está la IA débil, en el centro está la IA fuerte, y en el otro extremo se encuentra la IA súper fuerte.

### ANI

La primera categoría es la Inteligencia Artificial Estrecha (ANI, por sus siglas en inglés), también conocida como IA débil. La ANI está diseñada para tareas relativamente simples desde la perspectiva de la inteligencia artificial, como tecnologías básicas de reconocimiento de voz, autos autónomos y motores de recomendación de películas.

La ANI es relativamente fácil de desarrollar e implementar. Sin embargo, presenta dos limitaciones principales:

1. No puede aprender más allá de la tarea específica para la que fue diseñada.

2. No puede adaptarse a nuevas tareas o situaciones sin ser reprogramada.

Ejemplos de ANI incluyen Siri y Alexa.

### AGI

Pasando a la segunda categoría, tenemos la Inteligencia Artificial General (AGI), también conocida como IA fuerte. El objetivo de la AGI es imitar la inteligencia humana. Este tipo de IA puede aprender, razonar y adaptarse a nuevas situaciones por sí misma.

Actualmente no existe un ejemplo real de AGI, aunque abundan los ejemplos en películas de ciencia ficción como 2001: Odisea del Espacio.

La AGI representa, en términos científicos, algo tan revolucionario como la división del átomo.

Es importante considerar las películas de ciencia ficción cuando se habla de productos tecnológicos e inteligencia artificial, ya que influyen en el futuro de la inversión tecnológica, especialmente en el sector del capital de riesgo.
Por ejemplo, la película Minority Report de Steven Spielberg y Avatar influyeron en el desarrollo de tecnologías como la realidad virtual, aumentada, duplicación de pantalla y el concepto del metaverso.

Star Trek influyó en el mercado de los teléfonos móviles con los teléfonos plegables, y la película Her, protagonizada por Scarlett Johansson, inspiró el desarrollo de asistentes como Siri y Amazon Alexa.
Estos “emprendedores de ensueño” de Hollywood influyen en lo que Silicon Valley (u otros polos tecnológicos del mundo) desarrollarán en el futuro.
En 2001: Odisea del Espacio, el ejemplo de IA fuerte fue HAL, una computadora que podía pensar por sí misma. Curiosamente, si a H.A.L. le sumas una letra en cada letra, obtienes I.B.M.

Muchas personas están entusiasmadas con ChatGPT porque lo consideran lo más cercano a una AGI, ya que se le puede preguntar prácticamente cualquier cosa.

Si bien ChatGPT aún no es una AGI real, se está acercando rápidamente, y muchos inversores de capital de riesgo opinan que esta es la innovación tecnológica más asombrosa desde la invención de Internet.

Las ventajas de la AGI son:

1. Puede aprender y adaptarse a una amplia variedad de tareas. ChatGPT, por ejemplo, puede adaptarse en función de las preguntas que haces, pero aún no es una AGI verdadera.

2. Puede mejorar su rendimiento de forma autónoma.

Pero existen desafíos en el desarrollo de la AGI, entre ellos:

1. Requiere avances significativos en la tecnología de inteligencia artificial.

2. Plantea serias preocupaciones éticas.

Una vez que la AGI se convierta en una realidad, todas las empresas y gobiernos tendrán cargos dedicados a la ética de la IA, como un Departamento de Inteligencia Artificial y Ética o un Director de Ética en IA (Chief AI Ethics Officer).

Y no estamos nada lejos de ese escenario. Actualmente, existen varios centros de investigación altamente especializados trabajando activamente en el desarrollo de la IA, como OpenAI, DeepMind, y grandes empresas como IBM, Microsoft y Google.

Hoy en día, estas compañías ya deben contar con investigadores éticos externos que colaboran de forma continua para garantizar que el desarrollo de IA se realice de manera responsable.

### ASI

La tercera y última categoría de inteligencia artificial se llama Superinteligencia Artificial (ASI, por sus siglas en inglés), lo que significa que la IA sería más inteligente que los seres humanos en casi todos los aspectos. Aunque este concepto es especulativo, se debate ampliamente entre investigadores en IA y futuristas.

La ASI podría dar lugar a avances tecnológicos e innovaciones sin precedentes. Sin embargo, antes de que la Superinteligencia Artificial se vuelva realidad, será necesario lograr muchos descubrimientos clave en la investigación de IA.

Existen numerosos riesgos éticos y de otra índole si la IA llegara a superar el control humano desde el punto de vista de la ética, la seguridad y la regulación. Películas como Terminator, con el sistema ASI llamado Skynet, representan un ejemplo extremo de los problemas éticos que podría conllevar la ASI.

Incluso Warren Buffett advirtió que, una vez que abrimos la "caja de Pandora" de la IA, debemos tener cuidado. Es fundamental asegurarse de que los futuros productos basados en IA no obtengan demasiada autonomía.

## IA vs Machine Learning vs Deep Learning

Ahora que entendemos los 3 diferentes tipos de IA, discutamos las diferencias entre IA, Aprendizaje Automático y Aprendizaje Profundo, y veamos cómo se relacionan.

El **Aprendizaje Automático (Machine Learning)** es un subconjunto de la IA que consiste en desarrollar algoritmos para aprender patrones a partir de datos. Las máquinas aprenden de la experiencia, tal como lo hacen los humanos. En la sección 6 discutiremos cómo usar Excel para realizar Aprendizaje Automático usando Regresión Lineal y otras metodologías para encontrar patrones en los datos.

El **Aprendizaje Profundo (Deep Learning, DL)** es un subconjunto del Aprendizaje Automático (ML). El Aprendizaje Profundo utiliza Redes Neuronales Artificiales para modelar y resolver problemas complejos. Está inspirado en la estructura y función del cerebro humano.

El Aprendizaje Profundo es más complejo que el Aprendizaje Automático y requiere mejor hardware. Veamos algunos ejemplos de la vida real para entender las diferencias entre Aprendizaje Automático y Aprendizaje Profundo:

**Ejemplo 1: Sistemas de Recomendación de Películas.** Un sistema basado en IA podría usar reglas predefinidas para sugerir películas según el género o las preferencias del director. Un sistema de Aprendizaje Automático analizaría el comportamiento del usuario, como calificaciones e historial de visualización, para hacer recomendaciones personalizadas. Un sistema de Aprendizaje Profundo podría analizar datos no estructurados como resúmenes de las películas para ofrecer sugerencias aún más precisas.

**Ejemplo 2: Análisis de Texto.** Las técnicas de IA en texto podrían involucrar búsquedas por palabras clave y coincidencia de texto. Los algoritmos de Aprendizaje Automático pueden analizar datos de texto para descubrir patrones y relaciones, como análisis de sentimiento o modelado de temas. Los modelos de Aprendizaje Profundo, como las Redes Neuronales Recurrentes, pueden procesar patrones lingüísticos complejos para entender el contexto y generar respuestas avanzadas e inteligentes similares a las humanas.

**Ejemplo 3: Reconocimiento de Imágenes.** Los sistemas de reconocimiento de imágenes basados en IA usan algoritmos para identificar formas, colores y más en las imágenes. Los sistemas de Aprendizaje Automático usan técnicas de reconocimiento de patrones para clasificar objetos en imágenes, por ejemplo, identificar un perro y para crear imágenes. Los sistemas de Aprendizaje Profundo usan Redes Neuronales Convolucionales para analizar las imágenes a un nivel granular, haciendo que sean muy precisos en tareas como el reconocimiento facial.

En resumen, la IA es el concepto más amplio, que abarca el Aprendizaje Automático y el Aprendizaje Profundo. El Aprendizaje Automático es una forma de lograr IA entrenando modelos con datos, y el Aprendizaje Profundo usa redes neuronales para resolver problemas más complejos.

## Aprendizaje Supervisado / No Supervisado / Aprendizaje por Reforzamiento

Cuando trabajamos en una empresa y estamos analizando datos, nuestro jefe nos supervisa y nos indica qué columnas de datos usar, por ejemplo, una columna con fechas y otra con ventas.

Cuando somos bebés no tenemos estructura, porque gateamos y analizamos todo sin supervisión… como bebés no se nos enseña qué son las columnas de datos ni qué significan sus etiquetas.

**Categoría 1 de 3: Aprendizaje Supervisado (Supervised Learning)**

En el Aprendizaje Supervisado se le proporcionan a la computadora entradas (inputs), y luego la computadora usa aprendizaje automático para ofrecernos una salida (output).
La máquina puede usar el reconocimiento de patrones para crear una ecuación (por ejemplo, un Análisis de Regresión) con una entrada X para predecir un valor Y.

Ejemplos de análisis de regresión incluyen:

* Predicción de precios de acciones: es decir, predecir el precio futuro de una acción con base en datos históricos y otros factores.
* Pronóstico de demanda
* Optimización de precios
* Pronóstico de ventas
* Predicción de abandono de clientes (churn)
* Calificación crediticia en bancos
* Recomendación de productos
* Predicción del valor del cliente a lo largo del tiempo (Customer Lifetime Value)

Las tareas de aprendizaje supervisado son las más fáciles de resolver. Por eso, cuando las empresas quieren aplicar IA o Aprendizaje Automático a sus procesos, casi siempre comienzan preguntándose **¿podemos estructurar este problema como un problema supervisado?**

Por eso, actualmente la mayoría de los proyectos de Machine Learning en el mundo se pueden clasificar como tareas de aprendizaje supervisado.

**Ejemplo** https://appknnpy-scidata.streamlit.app/

**Categoría 2 de 3: Aprendizaje No Supervisado (Unsupervised Learning)**

El Aprendizaje No Supervisado es un subconjunto del Aprendizaje Automático donde se le da a la computadora una gran cantidad de datos sin etiquetar y se le pide que encuentre patrones o agrupamientos (clusters). Por ejemplo: identificar que ciertos tipos de clientes prefieren ciertos tipos de productos.

Los algoritmos de aprendizaje no supervisado trabajan con datos sin etiquetar, con el objetivo de descubrir patrones subyacentes o grupos dentro de los datos.

Este tipo de aprendizaje puede ser útil para tareas como:

* **Segmentación de clientes**, es decir, identificar qué tipos de clientes son más propensos a comprar ciertos productos.
* **Detección de anomalías** en transacciones financieras, como fraudes o el análisis de datos por parte de organismos como la Comisión de Bolsa y Valores (SEC) para detectar si los fondos de inversión están operando con información privilegiada.

**Ejemplo:** https://appkmeanspy-scidata.streamlit.app/



**Categoría 3 de 3: Aprendizaje por Reforzamiento (Reinforcement Learning)**

En el Aprendizaje por Reforzamiento, que es la tercera categoría de Machine Learning, un agente (como una computadora o robot) aprende a tomar decisiones interactuando con un entorno. El agente recibe retroalimentación en forma de recompensas o castigos, y su objetivo es maximizar la recompensa acumulada con el tiempo.

Aplicaciones potenciales en negocios incluyen:

* Optimización de decisiones en la cadena de suministro
* Personalización de la experiencia del cliente
* Desarrollo de estrategias de inversión financiera, donde la recompensa, por supuesto, es ganar la mayor cantidad de dinero posible sin asumir demasiado riesgo.

En la práctica, los modelos de Aprendizaje Automático se entrenan con datos históricos para hacer predicciones o tomar decisiones. Como líder empresarial, es crucial asegurarse de que los datos usados para entrenar sean representativos, precisos y estén libres de sesgos que puedan afectar el desempeño del modelo.

Finalmente, ten en cuenta que los modelos de Machine Learning pueden requerir actualizaciones y mantenimiento regular conforme se disponga de nuevos datos o cambie el contexto del negocio. Por ejemplo, YouTube, Google y Facebook cambian frecuentemente sus algoritmos, y cualquier proceso de Machine Learning basado en dichos algoritmos también podría necesitar ser actualizado.

## Visión por computadora

La Visión por Computador (o Visión Artificial) es un campo de la Inteligencia Artificial que permite a las computadoras y máquinas interpretar y comprender información visual del mundo, como imágenes, videos o transmisiones de cámaras en tiempo real. Es como enseñar a una computadora a "ver" y dar sentido al mundo visual, tal como lo hacen los humanos.

El objetivo principal de la visión por computador es encontrar diversas técnicas para manipular y extraer información de las imágenes, con el fin de obtener insights accionables para nuestro negocio. Con esto, podemos lograr cosas como clasificar objetos en una imagen, detectar rostros, identificar objetos en movimiento, peatones y otros elementos.

Un ejemplo increíble de una empresa que utiliza la Visión por Computador es Tesla. Tesla tiene uno de los mejores equipos de IA del mundo y sistemas de Visión por Computador desarrollados por ellos. En lugar de usar lidars y radares para navegar su entorno, los autos de Tesla utilizan únicamente 8 cámaras posicionadas alrededor del vehículo con fines de Visión Artificial.

Además de su aplicación en autos, Tesla emplea la Visión por Computador para optimizar sus fábricas, con el objetivo de acelerar la línea de producción y fabricar más vehículos. Por ejemplo, en muchas de sus fábricas, gran parte del trabajo lo realizan robots.

Exploremos algunas aplicaciones de la Visión por Computador y las industrias que las utilizan hoy en día:

Una de las aplicaciones más comunes es la clasificación de imágenes. Básicamente, creamos un sistema que, dada una imagen, predice lo que aparece en ella. Algunos usos incluyen blogs que detectan si un usuario ha subido contenido inapropiado o sistemas que reconocen el Lenguaje de Señas.

Otra aplicación es la detección y segmentación de objetos. En lugar de solo decir "esta es una imagen de un peatón", queremos saber exactamente dónde se encuentra dentro de la imagen. Imagina un auto autónomo: si usáramos solo clasificación de imágenes, el vehículo se detendría cada vez que apareciera un peatón en el cuadro. Pero si sabemos que el peatón está en la acera (y no en la calle), podemos avanzar con seguridad.

Un ejemplo más de segmentación de imágenes son los drones autónomos, que identifican y evitan obstáculos al entregar paquetes, o los robots en los almacenes de Amazon que empaquetan cajas.

Veamos cómo algunas industrias se han revolucionado gracias a la IA y la Visión por Computador:

**Redes Sociales:** Plataformas como TikTok, Instagram y Facebook analizan cada imagen que subimos para hacerla buscable.

**Sector Minorista (Retail):** Se usa para gestión de inventario, monitoreo de estantes, seguimiento del comportamiento de compradores, recomendación de productos y detección de robos mediante el reconocimiento de actividades sospechosas en tiempo real.

**Marketing y Publicidad:** Las empresas analizan el comportamiento y preferencias de los clientes mediante cámaras en tiendas o redes sociales, creando campañas más dirigidas y efectivas.

**Recursos Humanos:** Algoritmos de visión por computador analizan entrevistas en video, expresiones faciales y lenguaje corporal para tomar decisiones de contratación más informadas.
